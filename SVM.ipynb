```python
import csv
import pandas as pd
import numpy as np
import tensorflow as tf
import math
import random
import sklearn
```

```python
file = pd.read_csv('6000_normal.csv', header=None)
```

```python
file = file.drop(8000)
```

```python
len(file.index)
```

```python
file
```

```python
#data shuffle with index

idx = list(range(8000))
random.shuffle(idx)
split_ratio = 0.8
length = len(idx)
split = math.floor(length * split_ratio)

#split train, test data
train, test = file.loc[idx[:split]], file.loc[idx[split:]]
```

```python
#seperate data with labels (y_train means labels for x_train)
x_train, y_train = train.loc[:,1], train.loc[:,2]
x_test, y_test = test.loc[:,1], test.loc[:,2]
```

```python
#remove '\n'

for i in range(len(x_train)):
    value = x_train.iloc[i]
    value = value.split('\n')
    x_train.iloc[i] = value

for i in range(len(x_test)):
    value = x_test.iloc[i]
    value = value.split('\n')
    x_test.iloc[i] = value
```
